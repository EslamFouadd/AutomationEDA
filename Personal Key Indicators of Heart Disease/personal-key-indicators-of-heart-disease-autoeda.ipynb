# Data Cleaning and Preprocessing

```python
df = df.rename(columns = {
    "HeartDisease": "Y",
    "BMI": "X0",
    "Smoking": "X1",
    "AlcoholDrinking": "X2",
    "Stroke": "X3",
    "PhysicalHealth": "X4",
    "MentalHealth": "X5",
    "DiffWalking": "X6",
    "Sex": "X7",
    "AgeCategory": "X8",
    "Race": "X9",
    "Diabetic": "X10",
    "PhysicalActivity": "X11",
    "GenHealth": "X12",
    "SleepTime": "X13",
    "Asthma": "X14",
    "KidneyDisease": "X15",
    "SkinCancer": "X16"
})
```

```python
df = autoclean(df)
df.head()
```

## Outliers

```python
for column in df:
        plt.figure(figsize=(17,1))
        sns.boxplot(data = df, x = column)
```

```python
features = ["X0", "X4", "X5", "X13"]

for column in df[features]:
        plt.figure(figsize=(17,1))
        sns.boxplot(data = df, x = column)
```

```python
for col in df[features]:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    whisker_width = 1.5
    lower_whisker = q1 - (whisker_width * iqr)
    upper_whisker = q3 + whisker_width * iqr
    df[col] = np.where(df[col] > upper_whisker, upper_whisker, np.where(df[col] < lower_whisker, lower_whisker, df[col]))
```

```python
features = ["X0", "X4", "X5", "X13"]

for column in df[features]:
        plt.figure(figsize=(17,1))
        sns.boxplot(data = df, x = column)
```

## Skewness

```python
from sklearn.preprocessing import QuantileTransformer

qt = QuantileTransformer(output_distribution='normal')
```

```python
df["X0"] = qt.fit_transform(pd.DataFrame(df["X0"]))
pd.DataFrame(df["X0"]).plot.density(color='purple')
plt.show()
```

```python
df["X13"] = qt.fit_transform(pd.DataFrame(df["X13"]))
pd.DataFrame(df["X13"]).plot.density(color='purple')
plt.show()
```
